# Identificación de Títulos Faltantes

Este notebook tiene como objetivo identificar los títulos de películas presentes en el dataset principal, utilizado para realizar el clustering, que aún no han sido scrapeados. Para ello, se compara un dataset con los títulos ya procesados contra otro que contiene los títulos agrupados por características de clúster, permitiendo planificar el scraping en batch. 

El resultado final es una lista de títulos faltantes que será utilizada posteriormente en el notebook de web scraping.

```python
import pandas as pd
import numpy as np
```

```python
# Data_cluster
df_clustered = pd.read_csv("C:\\Users\\Usuario\\Desktop\\Netflix\\Raw_data\\df_clustered.csv",index_col=0)

# Batch
scrap_ultimo_1 = pd.read_csv("C:\\Users\\Usuario\\Desktop\\Netflix\\Raw_data\\scrap_ultimo_1.csv")
scrap_ultimo_2 = pd.read_csv("C:\\Users\\Usuario\\Desktop\\Netflix\\Raw_data\\scrap_ultimo_2.csv")
scrap_ultimo_3 = pd.read_csv("C:\\Users\\Usuario\\Desktop\\Netflix\\Raw_data\\scrap_ultimo_3.csv")
scrap_ultimo_4 = pd.read_csv("C:\\Users\\Usuario\\Desktop\\Netflix\\Raw_data\\scrap_ultimo_4.csv")
scrap_ultimo_5 = pd.read_csv("C:\\Users\\Usuario\\Desktop\\Netflix\\Raw_data\\scrap_ultimo_5.csv")
scrap_ultimo_6 = pd.read_csv("C:\\Users\\Usuario\\Desktop\\Netflix\\Raw_data\\scrap_ultimo_6.csv")
scrap_ultimo_7 = pd.read_csv("C:\\Users\\Usuario\\Desktop\\Netflix\\Raw_data\\scrap_ultimo_7.csv")
scrap_ultimo_8 = pd.read_csv("C:\\Users\\Usuario\\Desktop\\Netflix\\Raw_data\\scrap_ultimo_8.csv")
scrap_ultimo_9 = pd.read_csv("C:\\Users\\Usuario\\Desktop\\Netflix\\Raw_data\\scrap_ultimo_9.csv")
scrap_ultimo_10 = pd.read_csv("C:\\Users\\Usuario\\Desktop\\Netflix\\Raw_data\\scrap_ultimo_10.csv")
scrap_ultimo_11 = pd.read_csv("C:\\Users\\Usuario\\Desktop\\Netflix\\Raw_data\\scrap_ultimo_11.csv")
scrap_ultimo_12 = pd.read_csv("C:\\Users\\Usuario\\Desktop\\Netflix\\Raw_data\\scrap_ultimo_12.csv")
scrap_ultimo_13 = pd.read_csv("C:\\Users\\Usuario\\Desktop\\Netflix\\Raw_data\\scrap_ultimo_13.csv")
scrap_ultimo_14 = pd.read_csv("C:\\Users\\Usuario\\Desktop\\Netflix\\Raw_data\\scrap_ultimo_14.csv")
scrap_ultimo_15 = pd.read_csv("C:\\Users\\Usuario\\Desktop\\Netflix\\Raw_data\\scrap_ultimo_15.csv")
```

#### Concatenar los dataframes

```python
tablon = pd.concat([scrap_ultimo_1,scrap_ultimo_2,scrap_ultimo_3,scrap_ultimo_4,
                    scrap_ultimo_5,scrap_ultimo_6,scrap_ultimo_7,scrap_ultimo_8,
                    scrap_ultimo_9,scrap_ultimo_10,scrap_ultimo_11,scrap_ultimo_12,scrap_ultimo_13,scrap_ultimo_14,scrap_ultimo_15],axis=0)

titulos_tablon = tablon["title"].unique()
tablon
```

```python
# Borrar duplicados y mal scrapeados
tablon = tablon.dropna(subset=["title_comment", "comment"], how="all")
tablon.info()
```

### Lógica para obtener las películas faltan por scrapear.

```python
# Convertir las columnas a conjuntos de títulos únicos
titulos_cluster = set(df_clustered["title"].unique())
titulos_tablon = set(tablon["title"].unique())

# Encontrar los títulos que están en df_clustered pero no en tablon
titulos_faltantes = titulos_cluster - titulos_tablon

# Crear un DataFrame con los títulos faltantes
titulos_no_scrapeados = pd.DataFrame(list(titulos_faltantes), columns=["title"])

titulos_no_scrapeados["title_comment"] = pd.NA
titulos_no_scrapeados["comment"] = pd.NA
titulos_no_scrapeados
```

```python
tablon = pd.concat([tablon,titulos_no_scrapeados],axis=0)
```
