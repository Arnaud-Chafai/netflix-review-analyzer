```python
import os
from dotenv import load_dotenv
import pandas as pd
import openai
import random
```

#### Conexi칩n con la *API* de *OpenAI*

```python
# Cargar variables de entorno desde .env
load_dotenv()

# Obtener la clave desde el archivo .env
api_key = os.getenv("OPENAI_API_KEY")

# Inicializar el cliente de OpenAI
client = openai.OpenAI(api_key=api_key)
```

```python
df = pd.read_csv("C:\\Users\\Arnaud\\Desktop\\Netflix\\Raw_data\\tablon.csv")
```

#### Seleccionar un t칤tulo aleatorio para representar:  *`T칤tulo recomendado por el cluster`*

```python
# Seleccionar una pel칤cula aleatoria
random_index = random.randint(0, len(df) -1)
title = df.iloc[random_index]["title"]

# Filtrar las filas donde "title" coincide con "pelicula_predicha"
comentarios = df[df["title"] == title][["title_comment", "comment"]]
print(f"El t칤tulo de la pel칤cula seleccionada aleatoriamente es: ** {title} **")
comentarios
```

### Petici칩n a `**GPT-4o**`

```python
# Realizar una solicitud al modelo GPT
respuesta = client.chat.completions.create(
    model="gpt-4o",
    messages=[
        {
            "role": "user",
            "content": f"""
Based on the following user comments about the movie or series '{title}', provide a concise analysis in this format:

游땕 **Positive Consensus:**
- One sentence summarizing the most common positive feedback.

丘멆잺 **Negative Consensus:**
- One sentence summarizing the most common criticism.

游늷 **Summary:**
- A brief, 2-line overall perception of the movie/series, based on user feedback.

游눫 **User Comments:**
{comentarios}
"""
        }
    ]
)
```

```python
# Imprimir la respuesta del modelo
print(respuesta.choices[0].message.content.strip())
```
